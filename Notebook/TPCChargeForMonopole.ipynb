```c++
TF1 *fCh=new TF1 ("charge","[1]*x*x",1,5);

const  Double_t nEl=20.; //20.;   /// number of electrons/(MIP*cm), TPC-TRD p 151
const  Double_t gain= 6000.;  ///typical gas gain factor
const  Double_t fcoul= 1.6*1e-4;  ///femto coulomb per electron
const  Double_t dEdXrat= 4000;  /// ratio of dE/dX of monopole (g=1g_D and beta=0.9) and MIP - see dEdXMono notebook
const  Double_t len=1.;   //(cm) Integration length for charge ~ track length seen by one pad
const  Double_t expa=1.;   //TBD! charge dilution factor due to space charge expansion - see HighlyIonizingParticleDrift notebook


Double_t ch = nEl*len*gain*fcoul*dEdXrat*expa;

fCh->SetParameters(1,ch);

fCh->Draw();
fCh->GetHistogram()->GetYaxis()->SetTitle("fC");
fCh->GetHistogram()->GetXaxis()->SetTitle("g/g_{D}");
```
